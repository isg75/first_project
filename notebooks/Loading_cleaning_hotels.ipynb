```python
import pandas as pd

import pandas as pd

hotels_df = pd.read_csv(
    "/Users/levinschily/Desktop/IRONHACK/Week4/Project_Group/first_project/data/raw/opendatabcn_allotjament_hotels-csv.csv",
    encoding="utf-16",
    sep="\t",  
    engine="python"
)
hotels_df
```

```python
hotels_df.columns
```

```python
hotels_per_neighborhood = hotels_df.groupby(["addresses_neighborhood_name", "addresses_district_name"]).size().reset_index(name="hotel_count")

hotels_per_neighborhood["addresses_neighborhood_name"] = hotels_per_neighborhood["addresses_neighborhood_name"].str.lower()
hotels_per_neighborhood["addresses_district_name"] = hotels_per_neighborhood["addresses_district_name"].str.lower()

hotels_per_neighborhood = hotels_per_neighborhood.sort_values("hotel_count", ascending=False).reset_index(drop=True)

hotels_per_neighborhood
```

```python
hotels_per_neighborhood.to_csv("hotels_per_neighborhood.csv", index=False)
```
