```python
import pandas as pd

df = pd.read_csv("/Users/adamaskari/Downloads/charts.csv")
```

```python
top_50 = df[df["region"].isin(["United States", "United Kingdom", "Germany", "Spain"])]
```

```python
top_50 = (
    df[df["region"].isin(["United States", "United Kingdom", "Germany", "Spain"])]
      .assign(
          rank=df.groupby("region")["streams"]
                 .rank(method="first", ascending=False)
      )
      .query("rank <= 50")
      .sort_values(["region", "rank"])
)
display(top_50)
```
