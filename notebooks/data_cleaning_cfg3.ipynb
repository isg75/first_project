```python
import pandas as pd

# Assuming merged_games_data is already loaded as DataFrame
# Example to check and remove multiple genres from 'genres' column

# Define a function to check for genre presence and clean the genres
def update_genres(df, genre_list, column_name='genres'):
    """
    This function checks for the presence of specific genres in a column (e.g., 'genres') and appends 
    boolean columns indicating their presence. It also removes the genre from the original column.
    
    Parameters:
    - df: pandas DataFrame containing the data.
    - genre_list: List of genres to check and remove from the 'genres' column.
    - column_name: Name of the column to check and modify (default is 'genres').
    
    Returns:
    - The modified DataFrame with new boolean columns and cleaned 'genres' column.
    """
    for genre in genre_list:
        # Create a new column with boolean value indicating genre presence
        df[genre] = df[column_name].str.contains(genre, case=False, na=False)
        
        # Remove the genre from the 'genres' column if it exists
        df[column_name] = df[column_name].str.replace(genre, '', case=False, regex=False)
        
    # Clean up any unnecessary commas or spaces in the 'genres' column
    df[column_name] = df[column_name].str.replace(r'\s*,\s*', ', ', regex=True)  # Remove extra spaces around commas
    df[column_name] = df[column_name].str.replace(r'^\s*,|\s*,\s*$', '', regex=True)  # Remove leading/trailing commas

    return df

# List of genres to check and remove
genres_list = ['Indie', 'Early Access', 'Free to Play', 'Simulation', 'Sports', 'Action', 
               'Strategy', 'Adventure', 'RPG', 'Massively Multiplayer', 'Casual', 'Racing', 'Animation & Modeling', 'Design & Illustration', 'Education', 'Game Development']

# Apply the function to update the DataFrame
steam_data = update_genres(steam_data, genres_list)

# Check the updated DataFrame (showing just the first few rows and the new genre columns)
print(steam_data[['genres'] + genres_list].head())
```

```python
import pandas as pd

# Define a function to check for category presence and clean the categories
def update_categories(df, categories_list, column_name='categories'):
    """
    This function checks for the presence of specific categories in a column (e.g., 'categories') and appends 
    boolean columns indicating their presence. It also removes the category from the original column.
    
    Parameters:
    - df: pandas DataFrame containing the data.
    - categories_list: List of categories to check and remove from the 'categories' column.
    - column_name: Name of the column to check and modify (default is 'categories').
    
    Returns:
    - The modified DataFrame with new boolean columns and cleaned 'categories' column.
    """
    # Create boolean columns for each category
    for category in categories_list:
        # Create a new column with boolean value indicating category presence
        df[category] = df[column_name].str.contains(category, case=False, na=False)
    
    # Remove the specified categories from the 'categories' column
    for category in categories_list:
        df[column_name] = df[column_name].str.replace(category, '', case=False, regex=False)
    
    # Clean up any unnecessary commas or spaces in the 'categories' column
    df[column_name] = df[column_name].str.replace(r'\s*,\s*', ', ', regex=True)  # Remove extra spaces around commas
    df[column_name] = df[column_name].str.replace(r'^\s*,|\s*,\s*$', '', regex=True)  # Remove leading/trailing commas

    return df

# List of categories to check and remove
categories_list = ['Single-player', 'Multi-player', 'Steam Achievements', 'In-App Purchases', 'Partial Controller Support', 'Steam Cloud', 'Family Sharing', 
                   'Full controller support', 'Family Sharing', 'Steam Trading Cards', 'Captions available']

# Assuming 'steam_data' is already loaded as a DataFrame, apply the function
steam_data = update_categories(steam_data, categories_list)

# Check the updated DataFrame (showing just the first few rows and the new category columns)
print(steam_data[['categories'] + categories_list].head())
```
