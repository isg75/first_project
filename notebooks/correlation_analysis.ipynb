```python
import pandas as pd

df = pd.read_csv("/Users/levinschily/Desktop/IRONHACK/Week4/Project_Group/first_project/data/clean/combined_rest_and_hotel_count.csv")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


numeric_cols = ['rating', 'reviews', 'density', 'income', 'hotels_per_district']
corr_matrix = df[numeric_cols].corr()

print("Correlation with 'rating':")
print(corr_matrix['rating'].sort_values(ascending=False))

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```
