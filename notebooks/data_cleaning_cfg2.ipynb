```python
import pandas as pd

def update_categories_with_multiplayer_singleplayer(df, column_name='categories'):
    """
    This function appends two columns ('single-player' and 'multi-player') to the dataframe.
    It checks if the 'Single-player' or 'Multi-player' are present in the specified column 
    and appends the boolean values. It also removes these terms from the original column.

    Parameters:
    - df: pandas DataFrame containing the data.
    - column_name: Name of the column to check and modify (default is 'categories').

    Returns:
    - The modified DataFrame with two new boolean columns and cleaned 'categories' column.
    """
    # Check and append 'single-player' column
    df['single-player'] = df[column_name].str.contains('Single-player', case=False, na=False)

    # If 'Single-player' is in 'categories', remove it
    df[column_name] = df[column_name].str.replace('Single-player', '', case=False)

    # Check and append 'multi-player' column
    df['multi-player'] = df[column_name].str.contains('Multi-player', case=False, na=False)

    # If 'Multi-player' is in 'categories', remove it
    df[column_name] = df[column_name].str.replace('Multi-player', '', case=False)

    # Return the updated DataFrame
    return df


# Example usage
merged_games_data = pd.read_csv(r'C:\Users\Cecilia\Downloads\ironhack\coursework\group work\group project week3\first_project_cfg\data\raw\final_merged_steamdata.csv')

# Call the function
merged_games_data = update_categories_with_multiplayer_singleplayer(merged_games_data)

# Display the updated DataFrame
print(merged_games_data[['categories', 'single-player', 'multi-player']].head())
```

```python
import pandas as pd

# Assuming merged_games_data is already loaded as DataFrame
# Example to check and remove multiple genres from 'genres' column

# Define a function to check for genre presence and clean the genres
def update_genres(df, genre_list, column_name='genres'):
    """
    This function checks for the presence of specific genres in a column (e.g., 'genres') and appends 
    boolean columns indicating their presence. It also removes the genre from the original column.
    
    Parameters:
    - df: pandas DataFrame containing the data.
    - genre_list: List of genres to check and remove from the 'genres' column.
    - column_name: Name of the column to check and modify (default is 'genres').
    
    Returns:
    - The modified DataFrame with new boolean columns and cleaned 'genres' column.
    """
    for genre in genre_list:
        # Create a new column with boolean value indicating genre presence
        df[genre] = df[column_name].str.contains(genre, case=False, na=False)
        
        # Remove the genre from the 'genres' column if it exists
        df[column_name] = df[column_name].str.replace(genre, '', case=False, regex=False)
        
    # Clean up any unnecessary commas or spaces in the 'genres' column
    df[column_name] = df[column_name].str.replace(r'\s*,\s*', ', ', regex=True)  # Remove extra spaces around commas
    df[column_name] = df[column_name].str.replace(r'^\s*,|\s*,\s*$', '', regex=True)  # Remove leading/trailing commas

    return df

# List of genres to check and remove
genres_list = ['Indie', 'Early Access', 'Free to Play', 'Simulation', 'Sports', 'Action', 
               'Strategy', 'Adventure', 'RPG', 'Massively Multiplayer', 'Casual', 'Racing', 'JRPG']

# Apply the function to update the DataFrame
merged_games_data = update_genres(merged_games_data, genres_list)

# Check the updated DataFrame (showing just the first few rows and the new genre columns)
print(merged_games_data[['genres'] + genres_list].head())
```
