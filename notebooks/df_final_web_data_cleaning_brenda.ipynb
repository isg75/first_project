```python
#Load first dataset
import pandas as pd

pt1_df = pd.read_csv("../data/raw/df_final_web_data_pt_1.txt")
pt1_df
```

```python
#Load second dataset
pt2_df = pd.read_csv("../data/raw/df_final_web_data_pt_2.txt") 
pt2_df 
```

```python
#We merge these two dataframes prior to a comprehensive data analysis.
#Display the first "n" rows of the dataset
digital_footprints = pd.concat([pt1_df, pt2_df], ignore_index=True)
digital_footprints.head(20)
```

```python
#Retrievethe number of rows and columns in the dataframe
digital_footprints.shape
```

```python
#Display the data types of each column in the dataframe
digital_footprints.dtypes
```

```python
#Retrieve the unique data types present in the dataframe columns
list(set(digital_footprints.dtypes.tolist()))
```

```python
#Extract column names with numerical data types from the dataframe
digital_footprints.select_dtypes("number").columns
```

```python
#Count and sort the unique values for each numerical column in descending order
digital_footprints.select_dtypes("number").nunique().sort_values(ascending=False)
```

```python
#Checking for missing values and blank spaces (ROWS)
blank_rows = digital_footprints[digital_footprints['process_step'].isnull() | (digital_footprints['process_step'].str.strip() == '')]
print(blank_rows)
```

```python
#Checking missing values (COLUMNS)
digital_footprints.columns = digital_footprints.columns.str.strip()
print(digital_footprints.columns)
```

```python
#Check missing values of the whole dataframe
digital_footprints.isnull().sum()
```

```python
#Convert date_time column to datetime type
digital_footprints['date_time'] = pd.to_datetime(digital_footprints['date_time'])
digital_footprints.dtypes
```

```python
digital_footprints.to_csv('df_final_web_data_cleaned.csv', index=False)
```
