```python
import pandas as pd
import re


df_places = pd.read_csv("barcelona_places_data.csv")
income_df = pd.read_csv("2022_atles_renda_bruta_llar.csv")


def extract_district(address):
    match = re.search(r",\s*([^,]+),\s*080\d{3}", address)
    if match:
        return match.group(1).strip()
    return None

df_places["district"] = df_places["address"].apply(extract_district)
df_places["district"] = df_places["district"].str.strip().str.lower()

income_df.rename(columns={"Nom_Barri": "Neighbourhood"}, inplace=True)
income_df["Neighbourhood"] = income_df["Neighbourhood"].str.strip().str.lower()


income_grouped = income_df.groupby("Neighbourhood", as_index=False).agg({
    "Import_Renda_Bruta_€": "mean",
    "Seccio_Censal": "count"
})

merged_df = pd.merge(df_places, income_grouped, left_on="district", right_on="Neighbourhood", how="left")


merged_df.head()

df_places[["address", "district"]]

import re

def extract_district(address):
  
    match = re.search(r",\s*(.*?)\s*,\s*08\d{3}\s*Barcelona", address)
    if match:
        return match.group(1).strip()
    return None

df_places["district"] = df_places["address"].apply(extract_district)
df_places["district"] = df_places["district"].str.strip().str.lower()

df_places[["address", "district"]]
```
