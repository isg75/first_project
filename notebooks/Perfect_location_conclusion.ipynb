```python
import pandas as pd

con_df = pd.read_csv('combined_rest_and_hotel_count (1).csv')
con_df
```

```python
con_df = con_df.loc[:, ~con_df.columns.str.contains('^Unnamed')]
con_df
```

```python
grouped = con_df.groupby("district").agg({
    "rating": "mean",
    "reviews": "sum",
    "density": "mean",
    "income": "mean"
}).reset_index()

HIGH_INCOME = 25000
LOW_DENSITY = 100
HIGH_RATING = 4.5
HIGH_REVIEWS = 10000

top_restaurants = con_df.sort_values(['district', 'reviews'], ascending=[True, False]) \
                    .groupby('district').head(2)  

review_leaders = top_restaurants.groupby("district") \
    .apply(lambda g: ", ".join(f"{name} ({rating:.1f})" for name, rating in zip(g['name'], g['rating']))) \
    .reset_index(name="Review Leaders")

grouped = grouped.merge(review_leaders, on="district", how="left")

def classify(row):
    if row["density"] < 60:
        return "Too low density for casual traffic"
    elif row["income"] >= HIGH_INCOME and row["density"] < LOW_DENSITY:
        return "Low density; niche fine dining best"
    elif row["income"] >= HIGH_INCOME:
        return "High income, but quieter"
    elif row["rating"] >= HIGH_RATING:
        return "High potential"
    else:
        return "Good fallback option"

grouped["Suitability"] = grouped.apply(classify, axis=1)

grouped["income"] = grouped["income"].round(0).astype(int)
grouped["density"] = grouped["density"].round(0).astype(int)
grouped["rating"] = grouped["rating"].round(1)

print(grouped[["district", "income", "density", "Review Leaders", "Suitability"]])
```

```python
grouped = pd.DataFrame(grouped)
grouped
```

```python
import matplotlib.pyplot as plt

df_sorted_density = grouped.sort_values(by="density", ascending=False)

plt.figure(figsize=(10, 6))
plt.barh(df_sorted_density["district"], df_sorted_density["density"], color="#F08080")
plt.xlabel("Population Density (people per hectare)")
plt.title("Density by District")
plt.gca().invert_yaxis()
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
color_map = {
    'High potential': '#90EE90',
    'Good fallback option': '#ADD8E6',
    'High income, but quieter': '#FFD580',
    'Too low density for casual traffic': '#F08080'}


colors = grouped['Suitability'].map(color_map)

plt.figure(figsize=(10,6))
plt.bar(grouped['district'], grouped['income'], color=colors)
plt.xticks(rotation=45)
plt.title('Income by District (Suitability)')
plt.ylabel('Income (€)')
plt.tight_layout()
plt.show()
```
