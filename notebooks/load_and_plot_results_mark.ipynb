```python
import pandas as pd
```

```python
files = [
	"00_completion_rate_percentage",
	"01_avg_time_spent_per_step",
	"02_error_rate_percentage",
	"03_step_completion_rate_percentage"
]
experiments = ["both", "control", "test"]
```

```python
results = {}
for file in files:
	print(file)
	results[file] = pd.DataFrame()
	result = [pd.read_csv(f"../data/sql/output/{file}_{experiment}.csv").assign(source=experiment) for experiment in experiments]
	results[file] = pd.concat(result)
	results[file].reset_index(inplace=True)
	display(results[file])
```
