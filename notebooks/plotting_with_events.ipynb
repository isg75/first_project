```python
import yfinance as yf
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt

events = [
    { "name": "COVID", "date": datetime(2020, 3, 1) },
    { "name": "Something", "date": datetime(2021, 3, 1) },
    { "name": "Something Else", "date": datetime(2023, 3, 1) }
]

def plot_trend(ticker, start, end, events=[]):
    jpm = yf.Ticker(ticker)
    df = jpm.history(start=start, end=end)
    
    plt.figure(figsize=(12, 6))
    plt.plot(df.index, df["Close"], label=ticker + "Close Price")
    
    for event in events:
        plt.axvline(event["date"], color="red", linestyle="-", linewidth=1.5)
        plt.text(event["date"] + pd.Timedelta(days=10), df["Close"].max() * 0.9, event["name"], color="red", ha="left", va="bottom")
    
    plt.xlabel("Date")
    plt.ylabel("Closing Price")
    plt.title(ticker)


plot_trend("JPM", "2019-11-01", None, events)
plot_trend("BAC", "2019-11-01", None, events)
```
