```python
import pandas as pd

df = pd.read_csv("/Users/levinschily/Desktop/IRONHACK/Week4/Project_Group/first_project/data/clean/combined_rest_and_hotel_count.csv")
df
```

```python
df[["neighbourhood", "hotel_count"]].sort_values(by="hotel_count", ascending=False)
```

```python
df[["district_x", "hotel_count"]].groupby("district_x").sum().sort_values(by="hotel_count", ascending=False)
```

```python
# Plot district bar chart
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
district_df_sorted = df.sort_values(by="hotel_count", ascending=False)
plt.bar(district_df_sorted["district_x"], district_df_sorted["hotel_count"])
plt.title("Hotel Count by District")
plt.xlabel("District")
plt.ylabel("Number of Hotels")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plot neighbourhood bar chart
plt.figure(figsize=(12, 6))
neighbourhood_df_sorted = df.sort_values(by="hotel_count", ascending=False)
plt.bar(neighbourhood_df_sorted["neighbourhood"], neighbourhood_df_sorted["hotel_count"])
plt.title("Hotel Count by Neighbourhood")
plt.xlabel("Neighbourhood")
plt.ylabel("Number of Hotels")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
