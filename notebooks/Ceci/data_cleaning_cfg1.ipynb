```python
import pandas as pd
import numpy as np

imdb_games_data = pd.read_csv(r'C:\Users\Cecilia\Downloads\ironhack\coursework\group work\group project week3\first_project_cfg\data\raw\imdb-videogames.csv')

imdb_games_data.head()
```

```python
imdb_games_data.reset_index(drop=True, inplace=True)
imdb_games_data.columns = [c.lower().strip() for c in imdb_games_data.columns.values.tolist()]
imdb_games_data.head()
```

```python
# Filter the DataFrame for the year 2024
# Replace 'year' with the actual name of the year column in your DataFrame
imdb_games_2024 = imdb_games_data[imdb_games_data['year'] == 2024]

# Reset the index of the filtered DataFrame
imdb_games_2024.reset_index(drop=True, inplace=True)

# Display the filtered DataFrame
imdb_games_2024.head()
```

```python
# Drop the 'Unnamed: 0' column
imdb_games_2024.drop(columns=['unnamed: 0', 'url'], inplace=True)

# Display the filtered DataFrame without dropped columns
imdb_games_2024.head()
```
