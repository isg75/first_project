```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#url from Kaggle. Load the data from the file
data_path = '../data/raw/ai_job_dataset.csv'
df = pd.read_csv(data_path)
```

```python
#Combining and structuring data
#education_required, experience_level, salary_usd, required_skills, job_tittle, posting_date, company_location, benefits_score
new_df = pd.concat([df.education_required, df.experience_level, df.salary_usd, df.required_skills, df.job_title, df.posting_date, df.company_location, df.benefits_score ], axis=1)
display(new_df)
```

```python
#mode_val_education= df['education_required'].mode()
#print("The education required most common is (mode): ", mode_val_education)

#job_tittle = df['job_title'].mode()
#print("The job tittle most frequent is: ", job_tittle)

#Get the mode for each column
for column in ["education_required", "experience_level", "salary_usd", "required_skills", "job_title", "posting_date", "company_location", "benefits_score"]:
    
    mode = df[column].mode()
    print(f"Mode for {column}: {mode.tolist()}")
   

```

```python
# matplot and seaborn
import seaborn as sns
import matplotlib.pyplot as plt

#education_required, experience_level, salary_usd, required_skills, job_tittle, posting_date, company_location, benefits_score

#DISTRIBUTION OF SALARIES (USD)
plt.figure(figsize=(8, 5))
plt.hist(new_df['salary_usd'], bins=32, color='green', edgecolor='black')
plt.title('Distribution of Salaries (USD)')
plt.xlabel('Salary (USD)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

#BOXPLOT OF SALARY BY EXPERIENCE LEVEL
plt.figure(figsize=(10, 6))
sns.boxplot(x='experience_level', y='salary_usd', data=new_df)
plt.title('Salary Distribution by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Salary (USD)')
plt.show()
```

```python
#BAR CHART: AVERAGE SALARY BY EDUCATION LEVEL
avg_salary = new_df.groupby('education_required')['salary_usd'].mean()
plt.figure(figsize=(6,4))
avg_salary.plot(kind='bar', color='orange')
plt.title('Average Salary by Education Level')
plt.xlabel('Education Required')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
#JOB TITLE
# Count occurrences of each location
location_counts = new_df['job_title'].value_counts()

# Create pie chart
plt.figure(figsize=(6,6))
plt.pie(location_counts, labels=location_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('% JOB TITLE')

plt.axis('equal')  # Keep it circular
plt.show()

print(location_counts)

job_title_list = new_df["job_title"].tolist()
print(job_title_list)

sumatorio_job_title = len(job_title_list)
print("The sumatorio de job_titles, are: ", sumatorio_job_title)
```

```python
#yeas_experience & industry
#JOB TITLE
# Count occurrences of each location
industry_counts = df['industry'].value_counts()

# Create pie chart
plt.figure(figsize=(8, 6))
plt.pie(industry_counts, labels = industry_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Industries Sector % ')
plt.axis('equal')  # Keep it circular
plt.show()

print(industry_counts)

industry_counts_list = df["industry"].tolist()
print(industry_counts_list)

sumatorio_industries = len(industry_counts_list)
("The sumatorio de industries, are: ", sumatorio_industries)
```

```python
import matplotlib.pyplot as plt

# Agrupar por industria y calcular el promedio de aÃ±os de experiencia
industry_experience = df.groupby('industry')['years_experience'].mean().sort_values()

# Crear grÃ¡fico de barras
plt.figure(figsize=(8, 5))
industry_experience.plot(kind='bar', color='blue')
plt.title('Average Years of Experience Required by Industry')
plt.xlabel('Industry')
plt.ylabel('Average Years of Experience')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
new_df["skills_list"] = new_df["required_skills"].apply(lambda x: [skill.strip() for skill in x.split(",")])

# 2. Aplanar todas las listas en una sola lista
all_skills = [skill for sublist in new_df["skills_list"] for skill in sublist]

# 3. Contar cada habilidad
from collections import Counter
skill_counts = Counter(all_skills)

# 4. Obtener lista Ãºnica de habilidades
unique_skills = list(skill_counts.keys())

# 5. Mostrar resultados
print("ðŸ”¢ Conteo de habilidades:")
for skill, count in skill_counts.items():
    print(f"{skill}: {count}")

print("\nðŸ“‹ Lista de habilidades Ãºnicas:")
print(unique_skills)

sumatorio=len(unique_skills)
print("The total skills are: ", sumatorio)
```

```python
# Tidy skills by desdencendent frequency 
sorted_skill_counts = sorted(skill_counts.items(), key=lambda x: x[1], reverse=True)

print("ðŸ”¢ Skills count order by frecuency (descendente):")
for skill, count in sorted_skill_counts:
    print(f"{skill}: {count}")
```

```python
#SALARY TREND BY POSTING DATE
new_df['posting_date'] = pd.to_datetime(new_df['posting_date'])

# Group by date
daily_salary = new_df.groupby('posting_date')['salary_usd'].mean()

plt.figure(figsize=(12, 5))
plt.plot(daily_salary.index, daily_salary.values, color='green')
plt.title('Average Salary Over Time')
plt.xlabel('Date')
plt.ylabel('Average Salary (USD)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
