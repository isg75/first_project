```python
import pandas as pd

# Step 1: Load Data
df = pd.read_csv(r"C:\Users\Cecilia\Downloads\ironhack\coursework\group_work\group_project_week5_6\second_project\data\clean\combined_cleaned_data.csv")
```

```python
df.head()
```

```python
# Create bins for age groups
bins = [0, 30, 40, 50, 100] 
# Define corresponding labels for the age groups
labels = ['Under 30', '30-39', '40-49', '50 and above']

# Step 3: Create a new column 'age_group'
# Use pd.cut to categorize 'clnt_age' into the defined bins and assign the labels
df['age_group'] = pd.cut(df['clnt_age'], bins=bins, labels=labels, right=False)

df.head()
```
