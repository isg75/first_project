```python
df[["district", "hotels_per_district"]].drop_duplicates().sort_values(by="hotels_per_district", ascending=False)
```

```python
# Plot district bar chart
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
district_df_sorted = df.sort_values(by="hotels_per_district", ascending=False)
plt.bar(district_df_sorted["district"], district_df_sorted["hotels_per_district"])
plt.title("Hotel Count by District")
plt.xlabel("District")
plt.ylabel("Number of Hotels")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
