```python
mean_country_benefits = ai_job.groupby('company_location')['benefits_score'].mean().reset_index()
```

```python
sns.barplot(x='company_location', y='benefits_score',hue='company_location', data=mean_country_benefits,width=0.6,palette='viridis')
plt.axhline(y=ai_job['benefits_score'].mean(),color='orange',linestyle='-')
plt.title('Average of benefits_score per company_location')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
sns.barplot(x='benefits_score', y='company_location',hue='company_location', data=mean_country_benefits,width=0.6,palette='Set2')
plt.axvline(x=7.5, color='red', linestyle='--')
plt.title('Average of benefits_score per industry')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='company_location', y='benefits_score',hue='company_location',data=ai_job, palette='RdYlGn')
plt.xticks(rotation=45,ha='right')
plt.show()
```

```python
pivot = ai_job.pivot_table(index='company_location',columns='industry',values='benefits_score',aggfunc='mean')
plt.figure(figsize=(14,8))
sns.heatmap(pivot,annot=True,fmt='.1f',cmap='coolwarm')
plt.title('Average of benefits by country and industry')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
ai_job['industry'].nunique()
```

```python
mean_industry_benefit = ai_job.groupby('industry')['benefits_score'].mean().reset_index()
```

```python
sns.barplot(x='industry', y='benefits_score',hue='industry',data=mean_industry_benefit,width=0.6,palette='YlGnBu')
plt.axhline(y=7.5, color='r', linestyle='--')
plt.title('Average of benefits_score per industry')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 8)
plt.tight_layout()
plt.show()
```

```python
sns.barplot(x='benefits_score', y='industry',hue='benefits_score', data=mean_industry_benefit,width=0.6,palette='RdBu')
plt.axvline(x=7.5, color='red', linestyle='--')
plt.title('Average of benefits_score per industry')
plt.tight_layout()
plt.show()
```
