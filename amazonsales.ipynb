```python
# Show all column names
print(df.columns)
```

```python
# Drop unneeded columns
df.drop(['Order ID', 'Customer Name'], axis=1, inplace=True)
df.head()
```

```python
df['check_total'] = df['Price'] * df['Quantity']
df['is_correct'] = df['check_total'] == df['Total Sales']
print(df['is_correct'].value_counts())
```

```python
df['Product'] = df['Product'].str.strip().str.title()
print(df['Product'].unique())
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%y')
print(df['Date'].dtypes)
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
print(df['Status'].value_counts())
```

```python
payment_vs_status = df.groupby(['Payment Method', 'Status']).size().unstack().fillna(0)
print(payment_vs_status)
```

```python
location_vs_status = df.groupby(['Customer Location', 'Status']).size().unstack().fillna(0)
print(location_vs_status)
```

```python
triple_group = df.groupby(['Customer Location', 'Payment Method', 'Status']).size().unstack().fillna(0)
print(triple_group)
```

```python
# Show number of unique products and their frequencies
print("Unique products:", df['Product'].nunique())
print(df['Product'].value_counts())

# Analyze how product types relate to order status
product_status = df.groupby(['Product', 'Status']).size().unstack().fillna(0)
print(product_status)
```

```python
import matplotlib.pyplot as plt

# Plot bar chart
product_status.plot(kind='bar', stacked=True, figsize=(12, 6))

# Customisasi tampilan
plt.title('Order Status by Product')
plt.xlabel('Product')
plt.ylabel('Number of Orders')
plt.legend(title='Status')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plotting Payment Method vs Status
payment_vs_status.plot(kind='bar', stacked=True)

plt.title('Order Status by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Number of Orders')
plt.legend(title='Status')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plotting Customer Location vs Order Status
location_vs_status.plot(kind='bar', stacked=True)

plt.title('Order Status by Customer Location')
plt.xlabel('Customer Location')
plt.ylabel('Number of Orders')
plt.legend(title='Status')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Convert triple_group to a DataFrame for easier plotting
triple_df = triple_group.reset_index()

# Rename columns
triple_df.columns.name = None
print(triple_df.head())
```

```python
# Melt dataframe supaya bisa dipakai di seaborn
long_df = triple_df.melt(id_vars=['Customer Location', 'Payment Method'],
                         value_vars=['Cancelled', 'Completed', 'Pending'],
                         var_name='Status',
                         value_name='Count')

print(long_df.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set figure size
plt.figure(figsize=(14, 6))

# Create a bar plot for each payment method
sns.catplot(
    data=long_df,
    kind='bar',
    x='Customer Location',
    y='Count',
    hue='Status',
    col='Payment Method',
    height=5,
    aspect=1
)

plt.tight_layout()
plt.show()
```

```python
payment_methods = long_df['Payment Method'].unique()

for method in payment_methods:
    subset = long_df[long_df['Payment Method'] == method]

    plt.figure(figsize=(10, 5))
    sns.barplot(data=subset, x='Customer Location', y='Count', hue='Status')
    plt.title(f"Order Status by Customer Location ({method})")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
